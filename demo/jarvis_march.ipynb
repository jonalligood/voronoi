```python
from random import randint

from matplotlib import pyplot as plt

from geometry import Point
from jarvis_march import jarvis
```

```python
# Create a list of random sorted points
point_list = []
for i in range(100):
    point_list.append(Point(randint(0, 100), randint(0, 100)))

sorted_points = sorted(point_list, key=lambda p: (p.x, p.y))
```

```python
# Split the list into x and y lists for scatter plotting
x = []
y = []

for point in sorted_points:
    x.append(point.x)
    y.append(point.y)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
convex_hull = jarvis(sorted_points)
```

```python
hull_x = []
hull_y = []

for point in convex_hull:
    hull_x.append(point.x)
    hull_y.append(point.y)
```

```python
for i in range(0, len(hull_x)):
    plt.plot(hull_x[i:i+2], hull_y[i:i+2])
    
# Final line plot
final_x = [hull_x[-1], hull_x[0]]
final_y = [hull_y[-1], hull_y[0]]
plt.plot(final_x, final_y)
plt.scatter(x, y)
plt.show()
```
